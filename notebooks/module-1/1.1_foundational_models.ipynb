## Initialising and invoking a model

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(model="gpt-5-nano")
```

```python
response = model.invoke("What's the capital of the Moon?")

response
```

```python
print(response.content)
```

```python
from pprint import pprint

pprint(response.response_metadata)
```

## Customising your Model

```python
model = init_chat_model(
    model="gpt-5-nano",
    # Kwargs passed to the model:
    temperature=1.0
    #max_tokens=1000 per queries, limiting the number of tokens in the response, so that the model doesn't generate too long responses
    #timeout=30 time to wait for a response from the model before raising an error
    #max_retries=3 amount of times to retry the request if it fails
)

response = model.invoke("What's the capital of the Moon?")
print(response.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")

response = model.invoke("What's the capital of the Moon?")
print(response.content)
```

## Initialising and invoking an agent

```python
from langchain.agents import create_agent

agent = create_agent(model=model)
```

```python
agent = create_agent(model="claude-sonnet-4-5") #pick of of these models
```

```python
agent = create_agent("gpt-5-nano")
```

```python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?")]}
)
```

```python
from pprint import pprint

pprint(response)
```

```python
print(response['messages'][-1].content)
```

```python
from langchain.messages import AIMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?"),
    AIMessage(content="The capital of the Moon is Luna City."),
    HumanMessage(content="Interesting, tell me more about Luna City")]}
)

pprint(response)
```

## Streaming Output

```python
for token, metadata in agent.stream( #in roder to reduce latency and get the response as soon as possible, 
                                     #we can use the stream method to get the response in chunks
    {"messages": [HumanMessage(content="Tell me all about Luna City, the capital of the Moon")]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token
    
    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
```
