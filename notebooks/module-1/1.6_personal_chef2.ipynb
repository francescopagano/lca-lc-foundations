```python
from langchain.tools import tool
from typing import Dict, Any
from tavily import TavilyClient

tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:

    """Search the web for information"""

    return tavily_client.search(query)
```

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from psutil import users
from langgraph.checkpoint.memory import InMemorySaver  
from pprint import pprint
```

```python
system_prompt = """

You are a chef/baker, create a recipe for the user's requested dish.

Please keep to the below structure.

Name: The name of the recipe

Origin: Where it is from

Process: the process of making the dish

Be concise in the answer.

You have access to tools that can help you search the web for information. 
Use them if you need to find information about the dish.
"""

chef = create_agent(
    model="gpt-5-nano",
    system_prompt=system_prompt,
    checkpointer=InMemorySaver(),  
    tools=[web_search],
    
)
```

```python
question = HumanMessage(
    content="Give me a recipe for a lemon cake"
)
config = {"configurable": {"thread_id": "1"}}
```

```python
response = chef.invoke(
    {"messages": [question]},
    config
)

pprint(response)
```

```python
question = HumanMessage(content = "Can you tell me the most important ingredient for the recipe?")

response = chef.invoke(
    {"messages": [question]},
    config
)

pprint(response)
```
