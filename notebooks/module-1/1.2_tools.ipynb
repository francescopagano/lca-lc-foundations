## Tool definition

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.tools import tool

@tool
def square_root(x: float) -> float:
    """Calculate the square root of a number"""
    return x ** 0.5
```

```python
@tool("square_root")
def tool1(x: float) -> float:
    """Calculate the square root of a number"""
    return x ** 0.5
```

```python
@tool("square_root", description="Calculate the square root of a number")
def tool1(x: float) -> float:
    return x ** 0.5
```

```python
tool1.invoke({"x": 467}) # returns 21.602468994692867
```

## Adding to agents

```python
from langchain.agents import create_agent

agent = create_agent(
    model="gpt-5-nano",
    tools=[tool1], # list of tools to use
    system_prompt="You are an arithmetic wizard. Use your tools to calculate the square root and square of any number." 
    # system prompt for the agent, adding context about the tools and their usage
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="What is the square root of 467?")

response = agent.invoke(
    {"messages": [question]}
)

print(response['messages'][-1].content)
```

```python
from pprint import pprint

pprint(response['messages']) # prints the list of messages in the response, u can see that there is a tool call in the response, 
#which is the square root tool being called with the input 467
```
